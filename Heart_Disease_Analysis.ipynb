# 12. Find Gender Distribution according to the Target Column

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([0,1],['Male','Female'])
plt.show()
```

# 13. Check Age Distribution in the Dataset

# 14. Check Chest Pain Type

```python
sns.countplot(x='cp',hue='target',data=data)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.xticks(rotation =75)
plt.xticks()
```

# 15. Show Fasting Blood Sugar Distribution According to Target Variable

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Check Resting Blood Presurre Distribution

```python
sns.histplot(data['trestbps'])
```

# Compare Blood Pressure as per Sex Columns

```python
g = sns.FacetGrid(data,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['Male','Female'])
plt.show()
```

# 20.Plot Continous Variable

```python
cat_val = []
con_val = []

for columns in data.columns:
  if data[columns].nunique() <=10:
    cat_val.append(columns)
  else:
    con_val.append(columns)
```

```python
con_val
```

```python
data.hist(cat_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```

```python
data.hist(con_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
